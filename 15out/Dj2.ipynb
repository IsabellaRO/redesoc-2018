```python
# Calcular outdegree ponderado e indegree ponderado
contador = 0
diffeuclid = 10
while diffeuclid > 0.000001:
    for n in two.nodes:
        
#         Somando todos os in degrees de cada gênero e todos out de cada pessoa
#         Calculando In/Outdegree
        somai = 0
        somao = 0
        if contador == 0:
            for j in two.predecessors(n):
                somai += two.out_degree(j)
#             Definindo o indegree (de cada gênero) ponderado
#             if two.nodes[n]['Out-Degree ponderado'] == 0:
            two.nodes[n]['In-Degree ponderado'] = somai

            for i in two.successors(n):
                somao += two.in_degree(i)
#             Definindo o outdegree(de cada pessoa) ponderado
#             if two.nodes[n]['In-Degree ponderado'] == 0:
            two.nodes[n]['Out-Degree ponderado'] = somao

        else:
#             Recalculando e definindo o indegree (de cada gênero) e o outdegree(de cada pessoa) ponderados
            for j in two.predecessors(n):
                somai += two.nodes[j]['Out-Degree ponderado']
            if two.nodes[n]['Out-Degree ponderado'] == 0:
                two.nodes[n]['In-Degree ponderado'] = somai
            
            for i in two.successors(n):
                somao += two.nodes[i]['In-Degree ponderado']
            if two.nodes[n]['In-Degree ponderado'] == 0:
                two.nodes[n]['Out-Degree ponderado'] = somao
            

    table = pd.DataFrame({
        'Gênero/Aluno': [two.nodes[n]['label'] for n in two.nodes],
        'In-Degree': [two.in_degree(n) for n in two.nodes],
        'In-Degree ponderado': [two.nodes[n]['In-Degree ponderado'] for n in two.nodes],
        'Out-Degree': [two.out_degree(n) for n in two.nodes],
        'Out-Degree ponderado': [two.nodes[n]['Out-Degree ponderado'] for n in two.nodes]
    })

    #Somando todos os indegrees e todos os outdegrees para normalização
    somai = table['In-Degree ponderado'].sum()
    somao = table['Out-Degree ponderado'].sum()

    #Normalizando In/Outdegree
    normin = []
    normout = []
    for n in two.nodes:
        if two.nodes[n]['Out-Degree ponderado'] == 0:
            two.nodes[n]['In-Degree ponderado'] = two.nodes[n]['In-Degree ponderado']/somai
            normin.append(two.nodes[n]['In-Degree ponderado'])
        elif two.nodes[n]['In-Degree ponderado'] == 0:
            two.nodes[n]['Out-Degree ponderado'] = two.nodes[n]['Out-Degree ponderado']/somao
            normout.append(two.nodes[n]['Out-Degree ponderado'])

    #Recalculando In/Outdegree
    for n in two.nodes:
        somai = 0
        somao = 0
#         two.nodes[n]['In-Degree ponderado'] = 0
#         two.nodes[n]['Out-Degree ponderado'] = 0

        #Recalculando e definindo o indegree (de cada gênero) e o outdegree(de cada pessoa) ponderados
        for j in two.predecessors(n):
            somai += two.nodes[j]['Out-Degree ponderado']
        if two.nodes[n]['Out-Degree ponderado'] == 0:
            two.nodes[n]['In-Degree ponderado'] = somai
        for i in two.successors(n):
            somao += two.nodes[i]['In-Degree ponderado']
        if two.nodes[n]['In-Degree ponderado'] == 0:
            two.nodes[n]['Out-Degree ponderado'] = somao
    
    table = pd.DataFrame({
        'Gênero/Aluno': [two.nodes[n]['label'] for n in two.nodes],
        'In-Degree': [two.in_degree(n) for n in two.nodes],
        'In-Degree ponderado': [two.nodes[n]['In-Degree ponderado'] for n in two.nodes],
        'Out-Degree': [two.out_degree(n) for n in two.nodes],
        'Out-Degree ponderado': [two.nodes[n]['Out-Degree ponderado'] for n in two.nodes]
    })
    
    #Somando todos os indegrees e todos os outdegrees para normalização
    somai = table['In-Degree ponderado'].sum()
    somao = table['Out-Degree ponderado'].sum()

    #Renormalizando In/Outdegree
    normin2 = []
    normout2 = []
    for n in two.nodes:
        if two.nodes[n]['Out-Degree ponderado'] == 0:
            two.nodes[n]['In-Degree ponderado'] = two.nodes[n]['In-Degree ponderado']/somai
            normin2.append(two.nodes[n]['In-Degree ponderado'])
        elif two.nodes[n]['In-Degree ponderado'] == 0:
            two.nodes[n]['Out-Degree ponderado'] = two.nodes[n]['Out-Degree ponderado']/somao
            normout2.append(two.nodes[n]['Out-Degree ponderado'])

    #diferença euclidiana
    velhalista = normin + normout
    novalista = normin2 + normout2
    somaeuc = 0
    for k in range(len(velhalista)):
        somaeuc += (velhalista[k] - novalista[k])**2
    diffeuclid = math.sqrt(somaeuc)

    print('diffeuclid:', diffeuclid)
    
    contador += 1

print("contador:", contador)


table = pd.DataFrame({
        'Gênero/Aluno': [two.nodes[n]['label'] for n in two.nodes],
        'In-Degree': [two.in_degree(n) for n in two.nodes],
        'In-Degree ponderado': [two.nodes[n]['In-Degree ponderado'] for n in two.nodes],
        'Out-Degree': [two.out_degree(n) for n in two.nodes],
        'Out-Degree ponderado': [two.nodes[n]['Out-Degree ponderado'] for n in two.nodes]
})
# print(normout2)
```
