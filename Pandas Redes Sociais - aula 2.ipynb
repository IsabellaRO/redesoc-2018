```python
%matplotlib inline

import os
import matplotlib.pyplot as plt
from matplotlib import cm
import pandas as pd
import numpy as np

dados = pd.read_csv('cabreuva_limpa.csv', encoding = "ISO-8859-1")
dados
```

```python
novatab = dados.iloc[0:3430 , (114, 115)]
novatab.head()
```

```python
nova = dados.iloc[0:3430 , (114, 115)] #114 e 115
nova.raca = nova.raca.astype('category')
nova.raca.cat.categories = ('Branco', 'Negro', 'Moreno', 'Amarelo/Ascendência asiática', 'Indígena', 'Pardo, mulato, outros')
nova.religiao = nova.religiao.astype('category')
nova.religiao.cat.categories = ('Espírita', 'Evangélico', 'Católico', 'Religião afro-brasileira', 'Testemunha de Jeová, Mormon, Budista e outros', 'Sem religião', 'Ateu')
#1.branco, 2.negro, 3.moreno, 4.amarelo/asiatico, 5.indígena, 99. pardo, mulato, café com leite e outros
#1.Espírita, 2.Evangélico, 3.Católico, 4.Religião afro-brasileira, 5. Testemunha de Jeová, Mormon, Budista e outros, 6.sem religiao, 7. ateu
#nova[(nova.raca == 'Pardo, mulato, outros')]
#nova[(nova.religiao == 7)]
nova.head()
```

```python
nova.raca.value_counts()
```

```python
nova.religiao.value_counts()
```

```python
ct = pd.crosstab(nova.religiao, nova.raca)
ct
```

```python
soma_tabela = ct.sum().sum()
#ct/soma_tabela
ct*100/soma_tabela
```
